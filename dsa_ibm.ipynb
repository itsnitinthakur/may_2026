```python
# DSA top question solved
n=15
answer = []


for i in range(n):
    if (i+1)%3 == 0 and (i+1)%5 == 0:
        answer.append("FizzBuzz")
    elif (i+1)%3 == 0:
        answer.append("Fizz")
    elif (i+1)%5 == 0:
        answer.append("Buzz")
    else:
        answer.append(str(i+1))

print(answer)
```

```python
nums=[2,7,9,11]
target = 9
m = {}
n = len(nums)
for i in range(0,n):
    goal = target - nums[i]
    print(goal)
    if(goal in m):
        print([m[goal], i])
    print(nums[i], i)
    m[nums[i]] = i
```
