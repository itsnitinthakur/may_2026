```python
def lostFragments(sent, received):
    sent_words = sent.split()
    received_words = received.split()
    
    missing = []
    j = 0  # pointer for received
    
    for word in sent_words:
        if j < len(received_words) and word == received_words[j]:
            j += 1  # word was received, skip it
        else:
            missing.append(word)  # word was lost
    
    return missing

print(lostFragments("ddjjaff", ""))
```

```python
def minCost(numProjects, projectId, bid):
    min_bid = {}
    
    for pid, b in zip(projectId, bid):
        if pid not in min_bid or b < min_bid[pid]:
            min_bid[pid] = b
    
    if len(min_bid) < numProjects:
        return -1
    
    return sum(min_bid.values())
```

```python
def reorderBugs(bugs):
    freq = {}

    for bug in bugs:
        freq[bug] = freq.get(bug, 0) + 1

    result = []

    for bug in sorted(freq, key=lambda x: (freq[x], x)):
        result.extend([bug] * freq[bug])

    return result
```

```python
SELECT
    c.country_name,
    COUNT(*) AS total_domains
FROM domains d
JOIN countries c
    ON d.country_id = c.country_id
WHERE d.status = 'active'
GROUP BY c.country_name
ORDER BY c.country_name;
```
